```python
from matplotlib import pyplot as plt 

import numpy as np 
```

# Some nomenclature

Matplotlib figure and an axis 
- a figure is the full canvas we draw on -> can have multiple axes (e.g when we use twinx())
- an axis is the individual mathematical axes we use for plotting 

# Comparing the two ‘dialects’

```python
# data to be used for plotting 
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
          # both interfaces are supposed to give us two similar plots  
```

## pyplot Interface

```python
## dialect 1: pyplot interface 
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()

# purpose of pyplot interface: for quick and dirty plots, more for exploration 
```

## OO Interface

```python
## dialect 2: OO interface 
fig, ax = plt.subplots(nrows=1, ncols=1)  # start by creating a figure and axis 
ax.plot(x, cos_x, label='cos x')          # then create an axis 
ax.plot(x, sin_x, label='sin x')          # and another axis 
ax.legend()         

# purpose of OO interface: for more complex plots that demand control and finesse, impt for presentation of figures 
```

```python
## improving the above plot by using OO interface 
fig, ax = plt.subplots(nrows=1, ncols=1)  
ax.plot(x, cos_x, label='cos x')          
ax.plot(x, sin_x, label='sin x')          
ax.legend()  
ax.spines[['right', 'top']].set_visible(False)    # removing the top and right spines of the graph 
```

# What is OO ax, really?

```python
## creating 2 plots within the same figure 
# step 1: ask for 2 rows and 1 column using subplots() 
fig, ax = plt.subplots(nrows=2, ncols=1)    
```

```python
# step 2: adding in the details i.e plots and legends 
fig, ax = plt.subplots(nrows=2, ncols=1)   

ax[0].plot(x, cos_x, label = 'cos x')   # indexing [0] or [1] after ax indicates the graph we would like the plot to be on 
ax[1].plot(x, sin_x, label = 'sin x')

ax[0].legend()
ax[1].legend() 
```

# A complete OO example

```python
## an upgraded form of the previous plot, using more commands under the OO interface
# step 1: creating 2 empty axes with the correct dimensions 
fig, ax = plt.subplots(nrows=2, ncols=1,    # always start by creating the blank 2 empty axes 
                       figsize=(5, 5),      # changing the size of the figure 
                       sharex=True)         # ensuring that both plots share the x-axis 
```

```python
# step 2: adding the values of the 2 plots and the labels 
fig, ax = plt.subplots(nrows=2, ncols=1,    
                       figsize=(5, 5),      
                       sharex=True)         

                                            # since ax is a numpy array wwith 2 axes, we can use indexing 
ax[0].plot(x, cos_x, label=r'$\cos(x)$')    # x values is 'x', y values are 'cos x', plotting y = cos x graph, labelling 
ax[0].fill_between(x, 0, cos_x, alpha=.5)  # the start and end x values for plot 1 and the opacity of the shaded area 
ax[1].plot(x, sin_x, label=r'$\sin(x)$')    # x values is 'x', y values are 'sin x', plotting y = sin x graph, labelling 
ax[1].fill_between(x, 0, sin_x, alpha=.25)  # the start and end x values for plot 2 and the opacity of the shaded area 
```

```python
# step 3: adding more details, including the legends and the labels and gridlines 
fig, ax = plt.subplots(nrows=2, ncols=1,    
                       figsize=(5, 5),      
                       sharex=True)         

                                           
ax[0].plot(x, cos_x, label=r'$\cos(x)$')    
ax[0].fill_between(x, 0, cos_x, alpha=.5)  
ax[1].plot(x, sin_x, label=r'$\sin(x)$')    
ax[1].fill_between(x, 0, sin_x, alpha=.25)  

for a in ax:    # loop will run twice, once for x-axis and the other for the y-axis 
    a.legend()   
    a.grid(alpha=.25)  # setting opacity of the grid to be 25%
    a.set_ylabel('$y$', rotation=0)   # italicising axis labels and rotating them to in an orientation that is easy to read 

ax[1].set_xlabel('$x$')   # should be outside of loop since both plots share the same x-axis despite having different y-axis

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()      # prints out the 2 plots and the title of the figure closer together so they present better 
```

```python
## entire code of example 
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

      # Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

```python
   # step 1: create a figure and axes 

fig, ax = plt.subplots(nrows=2, ncols=2,   # 2 by 2 figure
                       figsize=(5, 5),   
                       sharex='col', sharey='row')   # tells python that 
                                                     # x axis is shared by plots in the same column while 
                                                     # y axis is shared by plots in the same row 
```

```python
    # step 2: setting titles and accessing each of the plots within a figure with mutiple plots 
 
ax[0, 0].set_title('Top Left')    # since this code is separated from the code creating the figure and the axes, 
ax[0, 1].set_title('Top Right')   # (con't) we dont see the titled plots printed  
ax[1, 0].set_title('Bottom Left') 
ax[1, 1].set_title('Bottom Right') # this works because the figure is 2x2. if the figure is 3x3, 

           # or, to make things easier to read and easier to tweak the title of the axes more efficiently 
top_left=ax[0, 0]
top_right=ax[0, 1]
bottom_left=ax[1, 0]
bottom_right=ax[1, 1] 

           # or 
top_left, top_right, bottom_left, bottom_right = ax.flatten()  # flatten() takes 2D assays and converts it into a 1D array
```

```python
     # step 3: including the horizontal lines 
              # reminder that for hlines() need to include the y value, and the start and end values of x 
top_left.hlines(y=1, xmin=0, xmax=5)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=5)
bottom_right.hlines(y=4, xmin=0, xmax=5)
```

## Sharing axes (sharex and sharey)

```python
# by sharing axes, we can make the plots more compact 
# how we choose to share axes is dependent on what we are trying to communicate with our plots 
```

```python
## without sharing of axes 

fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5))
                       #sharex='col', sharey='col')

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)
```

```python
## the sharing of axes 
# which option use depends on what we are trying to communicate with our plot 

   # option 1: True - makes all the axes use the same range 
   # option 2: col - use the same range for all the columns 
   # option 3: row - use the same range for all the rows 


# example 1: less suitable for the dsata we are plotting

fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row'   # all plots share the same scale of x-axis (0 to 5) but 
)                               # (con't) the 2 plots in the same row share the same y axis scale (3.0 to 4.0 and 1.0 to 2.0)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)
```

```python
# example 2: more suiable for the data we are plotting 

fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'  # every 2 plots in the same column will share the same x-axis scale (0 to 4 and 0 to 5)
)                               # every 2 plot in the same row will share the same y-axis scale (1.0 to 2.0 and 3.0 to 4.0)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)
```

## Accessing all axes

```python
## inefficient way: 
top_left.grid(alpha=.25)
top_right.grid(alpha=.25)
bottom_left.grid(alpha=.25)
bottom_right.grid(alpha=.25)
```

```python
## more efficient way - for loop 

for a in ax.flatten(): 
    a.grid(alpha = 0.25)
```

## Histograms (plt.hist())

### Data

### A quick helper function

```python
## defining a function that reads the numerical data in the 2 files 

def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)   # skip row because the first row is string 'basic_monthly_mean'

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)   # skip row because the first row is string 'basic_monthly_mean'

    return data   # here, data = { ['All'], ['NUS']} but only the numerical data 
```

```python
    # checking the content of the function and the data retained 
get_plot_data()  
```

### The histogram

```python
## using the defined function to plot a histogram 

plt.style.use('bmh')
data = get_plot_data()    # the function get_plot_data() READS the contents of the 2 files and returns it 

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])  # bins = 50 means for each set of data, there are 50 columns 
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')    # side note: rotation = 0 can be used to orientate the label horizontally 
plt.legend()    
```

```python
   # adjusting bins to see its effect 
plt.hist([data['All'], data['NUS']], bins=20, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')    # side note: rotation = 0 can be used to orientate the label horizontally 
plt.legend() 
```

## Scatter plots (plt.scatter())

here, we use the same data as in the histogram but also include x values using range() and len() to generate a list of x values for the earlier datasets

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items():   
    x = range(len(numbers))  
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Earning (S$)')
plt.legend()
```

## Bar charts (plt.bar()) or (plt.barh())

```python
## plotting a bar chart with vertical bars (plt.bar())

student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())   # majors = ['Life Sciences', 'Physics', 'Chemistry', 'Comp. Biology']
numbers = list(student_numbers.values())  # numbers = [ 14, 12, 8, 1]

plt.style.use('ggplot')
plt.bar(majors, numbers)  # syntax: plt.bar([x-axis data], [y-axis data])
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

```python
## plotting a bar chart with horizontal bars (plt.barh())

student_numbers = {'Egg': 4,
                   'Larva': 8,
                   'Pupa': 4,
                   'Adult': 20}
majors = list(student_numbers.keys())   # majors = ['Egg', 'Larva', 'Pupa', 'Adult']
numbers = list(student_numbers.values())  # numbers = [ 4, 8, 4, 20]

plt.style.use('ggplot')
plt.barh(majors, numbers)  # syntax: plt.bar([x-axis data], [y-axis data])
plt.xlabel('Days spent at different stages of life')
plt.ylabel('Stage of life cycle')
```

## Pie charts (plt.pie())

```python
student_numbers = {'Pea': 14,
                   'Rose': 12,
                   'Walnut': 8,
                   'Single': 1}

majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers,     # syntax: plt.pie(['y' values of each category], ['x' values or the categories])
        labels=majors,
        autopct='%1.2f%%',   # formatting the percentages, the number after the '.' represents the number of decimal places 
        startangle=-90       # the startangle affects the arrangement of the lines that cut the pie chart   
        )

plt.title('Number of chickens with each type of comb')
```
