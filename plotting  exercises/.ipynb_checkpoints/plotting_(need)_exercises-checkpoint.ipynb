```python
from matplotlib import pyplot as plt
```

Exercise 1 (Fill in the blanks)

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr = x_err, yerr = y_err,
             color = 'red', linestyle = 'dashed',
             marker = 'o', markerfacecolor = 'yellow',
             ecolor = 'black',capsize = 5,
             label = 'Y values with errors')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

Exercise 2 (A simple mathematical plot)

```python
import numpy as np 

x = np.linspace(0, 50, 200)
y = (np.e**(-x/10))*np.sin(x)   # change y limit to [-1, 1]

plt.plot(x, y, color = 'blue', linestyle = 'solid')  
plt.ylim([-1, 1])
plt.title('$y = e^{-x/10} \sin(x)$', fontsize = 20)  # equation of the function in the title has a font size of 20 
plt.grid(alpha = 0.25)  # grid with an opacity of 25%
plt.xlabel('x', color = 'darkorange', fontsize = 15)   # 'color' NOT 'colour'
plt.ylabel('y', color = 'darkorange', fontsize = 15)   # x and y axes labels in dark orange, font size 15 


# question - how should i rotate the y on the y axis to be easier to read as shown in the sp2273 exercise page 
```
